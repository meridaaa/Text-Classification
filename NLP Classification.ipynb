```python
import nltk
import numpy as np
import pandas as pd

filepath_dict = { 'v4 - Copy1.txt'}
df_list = []
  
for filepath in filepath_dict:
    df = pd.read_csv(filepath, names=['sentence', 'label'], sep='\t')
#     df['source'] = source 
    df_list.append(df)

df = pd.concat(df_list)
print(df)
```

```python
from __future__ import unicode_literals
from hazm import *
normalizer = Normalizer()
st=[]
s=[]


for i in df.sentence.values:
   normalizer.normalize(i)

tokenize=[]
for i in df.sentence.values:
       tokenize.append(word_tokenize(i))
# print(tokenize)
stemmer = Stemmer() 
for i in tokenize:
    for j in i:
          st.append(stemmer.stem(j))
    
# print(st)
lemmatizer = Lemmatizer()
l=[]

for i in st:
        l.append(lemmatizer.lemmatize(i))
print(l)
```

```python
import nltk
import numpy as np
# nltk.download('punkt')

stopword = open('stopword.txt', 'r', encoding='utf-8').read()

clean_sentence = [w for w in l if(not (w in stopword))]
for i in clean_sentence:
    n=i
    i = i.replace('\u200c', '')
    clean_sentence[clean_sentence.index(n)] = i
# print(int(y.shape))
print(clean_sentence)
o1=[]
o = []
y = df.label.values
for j in range(y.shape[0]):
          o1 = ' '.join(clean_sentence[:20])
          for i in clean_sentence[:20]:
              clean_sentence.remove(i)
          o.append(o1)  
```

```python
from sklearn.model_selection import train_test_split

sentences_train, sentences_test, y_train, y_test = train_test_split(
    o, y, test_size=0.25, random_state=0)
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
vectorizer.fit(sentences_train)

X_train = vectorizer.transform(sentences_train)
X_test  = vectorizer.transform(sentences_test)
print('vec X_train: \n', X_train.toarray())
print('\n')

print('vec X_test: \n',X_test.toarray())
```
